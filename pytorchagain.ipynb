# Simple linear Regression

`y = XW +b`

X is our input data, W is the weight, b is the bias.
y(y-hat) is our model's prediction or guess.

```python
# craeting a fake data
n=10
# each data have 1 input and 1 output function
D_in = 1
D_out = 1

# create our input data X
X = torch.randn(n, D_in)
print(X)
# Create our true target labels y by using the true W and b
# The true W is 2.0, the 'true' b is 1.0

true_W = torch.tensor([[2.0]])
true_b = torch.tensor(1.0)
y_true = X @ true_W + true_b + torch.randn(n, D_out) * 0.1
print(f'The output tensor is {y_true}')


## you see the part of `torch.randn(n+D_out)*0.1` is actually noise
```
