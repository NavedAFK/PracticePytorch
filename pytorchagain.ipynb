#building a graph

Goal: `z = x*y` where `y = a+b`

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)
x = torch.tensor(3.0, requires_grad=True)
y = a+b
z = y*x
print(z)
print(z.grad_fn)
```

```python
print(y.grad_fn)
```

```python
print(a.grad_fn)
```
